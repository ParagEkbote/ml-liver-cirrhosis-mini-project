```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/home/parag/ml-notes-and-code/data/sample2_dataset.csv")
```

```python
df
```

```python
%matplotlib inline
plt.scatter(df.age,df.weight)
plt.show()
```

```python
reg= linear_model.LinearRegression()
reg.fit(df[['age']],df.weight)
```

```python
predicted_value = reg.predict([[12]])[0]
print(f"Model prediction: {predicted_value}")
```

```python
plt.xlabel('Age')
plt.ylabel('Weight')
plt.scatter(df.age, df.weight,color='red')
plt.plot(df.age,reg.predict(df[['age']]))
plt.show()
```
