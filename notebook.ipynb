Decision Tree for salary_data.csv

```python
import pandas as pd
from sklearn.model_selection import LeaveOneOut, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

df = pd.read_csv("/home/parag/ml-notes-and-code/data/salary_data.csv")

target_col = df.columns[-1]
X = df.drop(target_col, axis=1)
y = df[target_col]

X = pd.get_dummies(X, drop_first=True)

if y.dtype == "object":
    y = pd.factorize(y)[0]

# ----------------------------
# 3. Leave-One-Out Cross Validation
# ----------------------------
loo = LeaveOneOut()
clf = DecisionTreeClassifier(max_depth=3, random_state=42)

scores = cross_val_score(clf, X, y, cv=loo)
print(f"ðŸŽ¯ Mean Accuracy (LOOCV): {scores.mean():.3f}")


```

```python
# ----------------------------
# 3. Leave-One-Out Cross Validation
# ----------------------------
loo = LeaveOneOut()
clf = DecisionTreeClassifier(max_depth=3, random_state=42)

scores = cross_val_score(clf, X, y, cv=loo)
print(f"ðŸŽ¯ Mean Accuracy (LOOCV): {scores.mean():.3f}")

# ----------------------------
# 4. Fit classifier on full dataset for visualization
# ----------------------------
clf.fit(X, y)

# ----------------------------
# 5. Visualize the tree
# ----------------------------
plt.figure(figsize=(15, 10))
tree.plot_tree(
    clf,
    filled=True,
    feature_names=X.columns,
    class_names=[str(cls) for cls in sorted(set(y))],
    rounded=True
)
plt.title("Decision Tree Visualization", fontsize=14)
plt.show()
```
